```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def taylorSeries(x, terms):
    approximation = 0
    for i in range(terms):
        term = ((-1)**i * x**(2*i + 1)) / math.factorial(2*i + 1)
        approximation += term
    return approximation
```

```python
x = np.arange(0, 10.1, 0.1)
```

```python
actualSin = np.sin(x)
```

```python
terms = [1, 2, 3, 4]
errors = {}
```

```python
for i in terms:
    approx = taylorSeries(x, i)
    error = np.abs(actualSin - approx)
    errors[i] = error
    plt.plot(x, error)
```

## the error grows significantly as x increases. A 4-term approximation is incredibly accurate near x=0, but it fails as you approach x=10
