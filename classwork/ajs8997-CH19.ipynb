```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

def euler_forward_stable(h, t_start, t_end, max_points=5000):
    total_steps = int((t_end - t_start) / h)
    
    # Calculate how often to save a data point to prevent memory overload
    save_interval = max(1, total_steps // max_points)
    
    t_saved = []
    y_saved = []
    
    t = t_start
    y = 0.0
    
    t_saved.append(t)
    y_saved.append(y)
    
    for i in range(1, total_steps + 1):
        dy_dt = -1000.0 * y + 3000.0 - 2000.0 * np.exp(-t)
        y = y + h * dy_dt
        t = t_start + i * h
        
        # Prevent the math from overflowing to infinity and crashing the plotter
        if y > 1e100: y = 1e100
        if y < -1e100: y = -1e100
        
        if i % save_interval == 0 or i == total_steps:
            t_saved.append(t)
            y_saved.append(y)
            
    return t_saved, y_saved

def analytical_solution(t):
    return 3 - 0.998 * np.exp(-1000 * t) - 2.002 * np.exp(-t)

h_values = [0.1, 0.01, 0.001, 0.00001, 0.0000001]
t_start = 0.0
t_end = 30.0

plt.figure(figsize=(10, 15))

for i, h in enumerate(h_values):
    t_num, y_num = euler_forward_stable(h, t_start, t_end)
    
    t_ana = np.linspace(t_start, t_end, 1000)
    y_ana = analytical_solution(t_ana)
    
    plt.subplot(len(h_values), 1, i + 1)
    plt.plot(t_ana, y_ana, 'k-', linewidth=2, label="Analytical Solution")
    plt.plot(t_num, y_num, 'r--', label=f"Euler Forward")
    
    plt.title(f"Step Size: h = {h}")
    plt.xlabel("t")
    plt.ylabel("y(t)")
    
    # Restrict y-axis for unstable step sizes to keep the graph readable
    if h >= 0.01:
        plt.ylim([-5, 10])
        
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```
