```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: np.sin(x)
a, b = 0, 2 * np.pi
N = 10             
x = np.linspace(a, b, N)
```

```python
h = (b - a) / (N - 1)
```

```python
exact_dx = np.cos(x)
```

```python
dx=(f(x+h)-f(x-h))/(2*h)
```

```python
forward_dx = (f(x+h) - f(x)) / h
```

```python
backward_dx = (f(x) - f(x-h)) / h
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Comparison of Numerical Differentiation Methods for f(x) = sin(x)")


axs[0, 0].plot(x, np.cos(x), color='black')
axs[0, 0].set_title('Exact Derivative: cos(x)')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel("f'(x)")


axs[0, 1].plot(x, np.cos(x), label='Exact cos(x)', color='black', linestyle='--')
axs[0, 1].plot(x, forward_dx, label='Forward Difference', color='blue')
axs[0, 1].set_title('Forward Difference')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel("f'(x)")
axs[0, 1].legend()


axs[1, 0].plot(x, np.cos(x), label='Exact cos(x)', color='black', linestyle='--')
axs[1, 0].plot(x, backward_dx, label='Backward Difference', color='red')
axs[1, 0].set_title('Backward Difference')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel("f'(x)")
axs[1, 0].legend()

axs[1, 1].plot(x, np.cos(x), label='Exact cos(x)', color='black', linestyle='--')
axs[1, 1].plot(x, dx, label='Central Difference', color='green')
axs[1, 1].set_title('Central Difference')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel("f'(x)")
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```
