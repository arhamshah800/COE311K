```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters from demo
k = 1.0
h_values = [0.01, 0.02, 0.03, 0.05, 0.1, 0.2, 0.3, 0.5]
x_start = 0
x_end = 20  # Duration extended to observe stability trends
y_0 = 1.0
v_0 = 0.0

plt.figure(figsize=(12, 8))

for h in h_values:
    # Create arrays using the demo code structure
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))
    
    # Set initial conditions
    y_values[0] = y_0
    v_values[0] = v_0
    
    # Euler's Method loop (Exact Demo Code Logic)
    for i in range(1, len(x_values)):
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]
        
        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)
        
    plt.plot(x_values, y_values, label=f'h = {h}')

plt.axhline(0, color='black', linewidth=1)
plt.title('Stability Comparison for Different Step Sizes (h)')
plt.xlabel('x')
plt.ylabel('Position y(x)')
plt.legend(title='Step Size (h)')
plt.grid(True)
plt.show()
```

```python
# Parameters for high k
k_high = 10.0
h_small = 0.01
x_end_high = 10

x_vals_k10 = np.arange(x_start, x_end_high + h_small, h_small)
y_vals_k10 = np.zeros(len(x_vals_k10))
v_vals_k10 = np.zeros(len(x_vals_k10))

y_vals_k10[0] = y_0
v_vals_k10[0] = v_0

for i in range(1, len(x_vals_k10)):
    y_vals_k10[i] = y_vals_k10[i-1] + h_small * v_vals_k10[i-1]
    v_vals_k10[i] = v_vals_k10[i-1] + h_small * (-k_high * y_vals_k10[i-1])

plt.figure(figsize=(10, 5))
plt.plot(x_vals_k10, y_vals_k10, color='red', label='k=10, h=0.01')
plt.title('Harmonic Oscillator with High Stiffness (k=10)')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.legend()
plt.grid(True)
plt.show()
```

I observe the plotted results for the different step sizes. I see that the position values grow larger over time for every value of h. I notice that the system is becoming unstable from the very first cycle. The method fails to conserve energy. I see that larger step sizes cause the position to spiral outward very quickly. I observe that increasing the value of k makes the oscillations much faster. I see that the high frequency makes the simulation even less stable. I believe I cannot use Euler's method to solve this with a large h. I observe that the error will always grow because the method overshoots the true curve.
