```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def monteCarlo (iterations):
    counter = 0
    for i in range(0,iterations):
        x = np.random.uniform(-1, 1)
        y = np.random.uniform(-1, 1)
        if (((x * x) + (y * y)) < 1):
            counter += 1
    myPi = 4*counter/iterations
    return myPi
```

```python
z = np.arange(10,10000,20)
values = []
for samples in z:
    y = monteCarlo(samples)
    values.append(y)
values = np.array(values)
abserror = np.abs(values - np.pi)
relerror = abserror/np.pi
plt.plot(z, abserror,label='Absolute Error')
plt.plot(z,relerror,label='Relative Error')

plt.legend() # Display the legend
plt.title("Monte Carlo Pi Approximation")
plt.xlabel("Number of Samples")
plt.ylabel("Error")
plt.show()
```
