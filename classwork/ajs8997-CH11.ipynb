```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Define known data points
x = np.array([0, 1, 2, 3, 4, 5]) # [cite: 285, 286, 287, 288, 289, 290]
y = np.array([0, 1, 0, 1, 0, 1]) # [cite: 291, 292, 293, 294, 295, 296]

# 1. Create a cubic spline interpolation (Natural Boundary Conditions)
CS = CubicSpline(x, y, bc_type='natural') # [cite: 299]

# 2. Create a global polynomial fit (Degree 5 perfectly interpolates 6 points)
# We ignore the rank warning that numpy might throw for this exact fit
with np.errstate(all='ignore'):
    poly_coeffs = np.polyfit(x, y, 5)

# Generate new x values for a smooth plot
x_interp = np.linspace(0, 5, 100) # [cite: 301]

# Calculate interpolated y values for both methods
y_spline = CS(x_interp) # [cite: 304]
y_poly = np.polyval(poly_coeffs, x_interp)

# Plot both on the same graph
plt.figure(figsize=(10, 5))
plt.plot(x, y, 'bo', label="Data points", zorder=5)
plt.plot(x_interp, y_spline, 'r-', label="Cubic Spline (Natural)")
plt.plot(x_interp, y_poly, 'g--', label="Polynomial Fit (Degree 5)")

plt.title("Challenge: Cubic Spline vs Polynomial Fitting")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
