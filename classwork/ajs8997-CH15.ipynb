```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def DerivativeFunc(X, Y):
    return -2 * Y
```

```python
XStart, XEnd = 0, 2
StepSizes = np.arange(0.01, 1.01, 0.05)
```

```python
BaselineStep = StepSizes[0]
BaselineNumSteps = int((XEnd - XStart) / BaselineStep)
BaselineX = np.linspace(XStart, XEnd, BaselineNumSteps + 1)
BaselineY = np.zeros(BaselineNumSteps + 1)
BaselineY[0] = 1
```

```python
for I in range(BaselineNumSteps):
    BaselineY[I + 1] = BaselineY[I] + BaselineStep * DerivativeFunc(BaselineX[I], BaselineY[I])
```

```python
plt.figure(figsize=(12, 8))
```

```python
for StepSize in StepSizes:
    NumSteps = int((XEnd - XStart) / StepSize)
    XValues = np.linspace(XStart, XEnd, NumSteps + 1)
    YValues = np.zeros(NumSteps + 1)
    
    YValues[0] = 1  
    
    for I in range(NumSteps):
        YValues[I + 1] = YValues[I] + StepSize * DerivativeFunc(XValues[I], YValues[I])
        
    BaselineComparisonY = np.interp(XValues, BaselineX, BaselineY)
    PointErrors = np.abs(YValues - BaselineComparisonY)
    TotalError = np.sum(PointErrors)
    
    print(f"--- Run with StepSize = {StepSize:.2f} ---")
    print(f"Point Errors: {np.round(PointErrors, 4)}")
    print(f"Total Accumulated Error: {TotalError:.4f}\n")
    
    plt.plot(XValues, YValues, label=f"h={StepSize:.2f}", alpha=0.7)

plt.xlabel('X')
plt.ylabel('Y')
plt.title("Approximation of dy/dx = -2y using Euler's Method (Varying Step Sizes)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
